```python
# Question 5   
# create a custom function that converts temperatures from F to C or C to F
# the function should take 2 arguments, the temperature and the unit it is currently in.
# use an if statement to handle each of the units
# return the converted value and the new unit
# test the function on these values below.

def convertTemps(temp, unit):
    if unit == "F":
        celsiusTemp = (temp - 32) * 5 / 9
        print(f"{temp}F converted to celsius is ", round(celsiusTemp, 2), "C")
    elif unit == "C":
        fahrenheitTemp = temp * 9 / 5 + 32
        print(f"{temp}C converted to fahrenheit is ", round(fahrenheitTemp, 2), "C")
    else:
        return "Invalid unit"

# (a) 0F

convertTemps(0, "F")

# (b) 0C

convertTemps(0, "C")

# (c) 96F

convertTemps(96, "F")

# (d) 100C

convertTemps(100, "C")

 
```
