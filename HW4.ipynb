```python
# Question 1  
# Use regex to find all the words in the lorem string that include the digit '7' 

#Setting sevenMatch to find all the words with the number 7 in them regardless of what comes before or after
sevenMatch = re.findall(r'\b\w*7\w*\b', lorem)
print(f"The words that contain number 7 include: {sevenMatch}")
```

```python
# Question 2 
# Use regex to return all the words with a capitalized first letter

#Finding all the words that start with a capital letter regardless of what comes after it
capLetters = re.findall(r'\b[A-Z]\w*\b', lorem)
print(f"The words with capitalized letters include: {capLetters}")
print(len(capLetters))
```

```python
# Question 6
# The US National Parks system has webpages for each park
# start here https://www.nps.gov/bibe/index.htm
# and find the page for 'Birds of Big Bend'
# use that url to scrape the contents of the page
# and return all the photo captions in a list

#Setting the target URL 
url = "https://www.nps.gov/bibe/learn/nature/birds-of-big-bend.htm"
page = requests.get(url)

#Using beautifulsoup to parse the HTML code of the page
birdsOfBigBend = BeautifulSoup(page.content, 'html.parser')

#Isolating any parts of the HTML code that has figcaption in it as a html tag
photoCaptions = birdsOfBigBend.find_all('figcaption')

#Creating the empty list of photo captions
listOfPhotoCaptions = []

#For each caption in photoCaptions append them to the list
for i, caption in enumerate(photoCaptions):
    listOfPhotoCaptions.append(caption.text)

print("All of the photo captions include:")    
for i in listOfPhotoCaptions:
    print(i)
```

```python
# Question 7  
# There is a mailing address and phone number on the home page
# find and print that information as simply as you can

streetAddress = birdsOfBigBend.find("span", itemprop="streetAddress")
addressLocality = birdsOfBigBend.find("span", itemprop="addressLocality")
addressRegion = birdsOfBigBend.find("span", itemprop="addressRegion")
postalCode = birdsOfBigBend.find("span", itemprop="postalCode")
phoneNumber = birdsOfBigBend.find("span", itemprop="telephone")

print(f"The address is: {streetAddress.text} {addressLocality.text} {addressRegion.text} {postalCode.text} \nThe phone number is: {phoneNumber.text}")
```
