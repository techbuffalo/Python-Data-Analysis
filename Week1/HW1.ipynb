```python
# RUN THIS CELL FIRST

# import package
import numpy as np 
import pandas as pd 
```

```python
# RUN THIS CELL TO ANSWER THE QUESTIONS BELOW
# DO NOT MAKE ANY CHANGES TO THIS CELL

# create a 250x250 np matrix
np.random.seed(412)
matrix = np.random.uniform(0.001, 100, (250, 250))
matrix = np.round(matrix, 3)
```

```python
# Question 1  
# Find and print the minimum and maximum values in the 200x200 matrix
# created in the cell above and print the matrix location for each value

# find the minimum value and its location

minimumLocation = np.argmin(matrix)
minimumValue = np.min(matrix)
# find the maximum value and its location
maximumLocation = np.argmax(matrix)
maximumValue = np.max(matrix)
# print the results
print('Minimum value and location:', minimumValue, minimumLocation)
print('Maximum value and location:', maximumValue, maximumLocation)
```

```python
# Question 2  
# Count the number of values <= 1 and print that value
# and express that as a percentage of the total values in the matrix
# rounded to 3 decimal places
underOne = matrix[matrix < 1]
percentUnderOne = (len(underOne) / matrix.size) * 100

print(np.round(percentUnderOne, decimals=3))
```

```python
# Question 3  
# Remove all the values <= 1 and print the mean and median values
# for that reduced matrix rounded to 3 decimal places

# filter values > 1
greaterOne = matrix[matrix > 1]
# calculate the mean and median values for matrix values > 1
matrixMean = np.mean(greaterOne)
matrixMedian = np.median(greaterOne)
# print the results
print("The mean of the matrix with numbers greater than one is: ", np.round(matrixMean, decimals=3))
print("The median of the matrix with values greater than one is: ", matrixMedian)
```

```python
# Question 4  
# Use pandas to read the 'charleston_temps.csv' file
# NOTE the first few rows are notes and not part of the data
# create the code to extract ONLY the dataset
# starting at the header row
# use head() to check the new dataframe
dfCharleston = pd.read_csv('charleston_temps.csv', skiprows=4)
print(dfCharleston.head())
```

```python
# Question 7  
# Use pandas to read in the 'chocolate.csv' dataset 
# and use head() to look at the column headers and first few rows
dfChoclate = pd.read_csv('chocolate.csv')
dfChoclate.head()
```

```python
# Question 8  
# What are the earliest and latest Review Dates?
# also calculate the number of years this covers
reviewDate = dfChoclate.sort_values('Review Date')
print('The earliest review date was the year: ', reviewDate['Review Date'].min())
print('The latest review date was the year: ', reviewDate['Review Date'].max())
```

```python
# Question 9  
# Use pandas groupby to count the Companies rated for each Year
companyRating = dfChoclate.groupby("Review Date")[["Company"]].count()
print(companyRating)
# Complete the text string below by changing <your answer here>:
""" 
The year with the most number of entries was 2016.
"""
```

```python
# Question 10  
# Use pandas groupby to determine the mean Company for each Review Year
# and return the top 10 sorted in descending order with their ratings values
meanRating = dfChoclate.groupby(["Review Date", "Company"])["Rating"].mean()
sortedMeanRating = meanRating.head(10).sort_values(ascending=[False])
print(sortedMeanRating)
```
