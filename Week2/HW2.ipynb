```python
# Question 2  
# Using the 'bank_churn.csv' dataset, recreate the barchart 'bank_churn_barplot.png'.  
# the bars are colored orange.
df_bankChurn = pd.read_csv("Week2Data/bank_churn.csv")
educationLevels = df_bankChurn['education'].value_counts()

#Adjustments to match the x and y axis with the example and changed the width so the x-axis labels don't overrun
plt.figure(figsize=(10,5))
plt.ylabel('count')
plt.xlabel('education Level')
plt.title('count of unique instances in education column')

#Showing the graph
plt.bar(educationLevels.index, educationLevels.values, color='orange')

#Saving the graph to a picture
plt.savefig('bankChurnBarplot.png')
```

```python
# Question 3 
# Also using the 'bank_churn.csv' file, recreate the histogram 'bank_churn_age_hist.png'.  
# the bars are colored light blue.

#Set the figure size and how to display the x-axis
fig, ax = plt.subplots(figsize=(10,6))
ax.hist(df_bankChurn['age'], color='lightblue', bins = 20, edgecolor='black')

#Setting the labels of the x-axis and the tile
ax.set_xlabel('age')
ax.set_title('Histogram of age with mean in dashed red and median in dashed green')

#Setting the mean and median dashed lines on the graph
ax.axvline(df_bankChurn['age'].mean(), color='red', linestyle='dashed')
ax.axvline(df_bankChurn['age'].median(), color='green', linestyle='dashed')

#Save the graph to a picture
fig.savefig('bankChurnHistogram.png')
```

```python
# Question 4  
# Using the 'OH_riverwater.csv' dataset, recrete the time trend graph 'river_time-trend.png'
# the line color is green.

df_riverWater = pd.read_csv("Week2Data/OH_riverwater.csv")

fig, ax = plt.subplots(figsize=(15,6))

ax.plot(df_riverWater['date'], df_riverWater['alkalinity'], color='green')

#Setting the lines based on the y-axis to display the min, max, and mean
ax.axhline(df_riverWater['alkalinity'].mean(), color='red', linestyle='dashed')
ax.axhline(df_riverWater['alkalinity'].min(), color='black', linestyle='dashed')
ax.axhline(df_riverWater['alkalinity'].max(), color='black', linestyle='dashed')

#Setting the x-axis to show vertically and labeling the x, y, and title
ax.tick_params(axis='x', rotation=90)
ax.set_xlabel('date')
ax.set_ylabel('alkalinity')
ax.set_title('time trend of alkalinity \n min and max in dashed black and mean in red')

#Save the graph to an image
plt.savefig('riverTimeTrend.png')
```

```python
# Question 5  
# Also using the 'OH_riverwater.csv' file, recreate the scatter plot 'river_scatter.png'
# the points are colored gray.

fig, ax = plt.subplots(figsize=(10,6))

ax.scatter(df_riverWater['alkalinity'], df_riverWater['hardness'], color='gray', marker='x')

#Setting the x, y, and title of the graph
ax.set_xlabel('alkalinity')
ax.set_ylabel('hardness')
ax.set_title('scatter plot of alkalinity and hardness')

#Setting the x and y axis ranges
ax.set_ylim(60 , 200)
ax.set_xlim(40, 100)

#Set the y axis to only show the ticks from 80 -> 200
yticks = np.arange(80, 201, 20)
ax.set_yticks(yticks)


plt.savefig('riverScatter.png')
```
